### Method vs. Attribute in Pandas

In object-oriented programming (which pandas is built upon), objects (like your DataFrame `df`) have both **attributes** and **methods**.

*   **Attributes**: These are characteristics or properties of an object. They store data or provide information about the object's state. You access them directly without parentheses.
    *   **Example**: `df.shape` is an attribute that returns a tuple representing the dimensions (rows, columns) of the DataFrame. `df.index` or `df.columns` are also attributes.

*   **Methods**: These are functions that belong to an object. They perform actions, operations, or calculations on the object's data. You call them using parentheses `()` (even if they don't take any arguments).
    *   **Example**: `df.head()` is a method that returns the first `n` rows of the DataFrame. `df.sample()` and `df.info()` are also methods.

# Filtering Dataframe

- df[condition]

```python
# Return student with mark >=90
df[df["mark"]>=90]
```

```python
# Return class Five
df[df["class"]=="Five"]
# first df[] is used for slicing
# second df[] is criteria
```

```python
# Return class not Five
df[~(df["class"]=="Five")]
# first df[] is used for slicing
# second df[] is criteria
```

```python
# Return class not Five
df[df["class"]!="Five"]
# first df[] is used for slicing
# second df[] is criteria
```

```python
# Return female student at class Four
df[(df["gender"]=="female") & (df["class"]=="Four")]
```

- blue [] is slicing
- brown [] is condition

```python
# Return female student or student class Four
df[(df["gender"]=="female") | (df["class"]=="Four")]
```

```python
# Return student with mark >=90
# df[df["mark"]>=90]
df.loc[df["mark"]>=90, :]
# : means return entire range
```

```python
# df.iloc[df["mark"]>=90, :]
```

```python
# Return female student at class Four
# df[(df["gender"]=="female") & (df["class"]=="Four")]
df.loc[(df["gender"]=="female") & (df["class"]=="Four"), :]
```
